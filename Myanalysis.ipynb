```python
import pandas as pd
import numpy as np
```

```python
all_data = pd.read_csv('Sales_Data/all_data.csv')
all_data.head()
```

# Removing nan

```python
 pd.isna(all_data).sum()
```

```python
pd.isna(all_data)
```

```python
all_data.dropna(inplace=True)
```

```python
all_data
```

### ADDING MONTH COLUMN

```python
all_data['Month']=all_data['Order Date'].apply(lambda x:x[0:2])   # .str[0:2]
```

```python
all_data.head()
```

### Making month column integer

```python
all_data=all_data[all_data['Month']!='Or']   # first removing or from month column
```

```python
all_data['Month']=all_data['Month'].apply(lambda x:int(x))
```

```python
all_data['Quantity Ordered']=all_data['Quantity Ordered'].apply(lambda x:int(x))
```

```python
all_data['Price Each']=all_data['Price Each'].apply(lambda x:float(x))
```

### Now that the month is integer 

```python
all_data['Sales']=all_data['Quantity Ordered']*all_data['Price Each']
all_data.head()   # Making sales column
```

### Find best month of sales?how much was earned that month?

```python
all_data.groupby('Month').sum()['Sales'].sort_values()   # december
```

## Which city has highest sales

```python
all_data['Purchase Address']
```

```python
def slicer(var):
    a=var.find(',')
    varnew=var[a+1:]
    b = varnew.find(',')
    return varnew[:b]
```

### all_data["City"]=all_data['Purchase Address'].apply(slicer)   or
### apply(lambda x : x.split(,)[1])

#### In the above example we have taken city alone but multiples states can have different city with same name ,so city name
#### must have state also. Now i will do it with both methods using lambda

```python
all_data['City']=all_data['Purchase Address'].apply(lambda x: slicer(x) + ' ' + x.split(',')[2].split(' ')[1] )
```

```python
all_data.head()
```

```python
Sales_city=all_data.groupby('City').sum()

City_index=Sales_city.index
```

```python
fig,axes = plt.subplots(figsize=(10,6))

axes.bar(City_index,Sales_city['Sales'])
axes.set_xticks('City_index',minor=True)

plt.xticks(rotation='vertical')
axes.set_xlabel('City')
axes.set_ylabel('Sales')
axes.ticklabel_format(axis='y',style='plain')
axes.grid(alpha=.5,linestyle="dashed")
#  Error-axes.ticklabel_format(style='plain') because x axis cannot be set as palin or scientific
```

# Q-4 What products are most often sold together?

```python
all_data.head(10)
```

```python
df=all_data[['Order ID','Product']]
df.head()
```

```python
df2=pd.merge(df,df,how="inner",on='Order ID')
df2.head(10)
```

```python
df3=df2[df2['Product_x']!= df2['Product_y']]
df3  
```

## Sequencer lagane ki jarurat ni padegi if you find duplicates

#### Trying to remove duplicates

```python
df3.head()
```

```python
#df3=df3[df3.duplicated(subset='Order ID')]
```

```python
df3['Products_mix']=df3['Product_x']+";"+df3['Product_y']
df3.head(6)
```

```python
def sequencer(str1):
    arr=str1.split(';')
    arr.sort()
    arr=' '.join(arr)
    return  arr
    
    
```

```python
df3['Products_mix']=df3['Products_mix'].apply(sequencer)
df3.head()
```

## Now we will check duplicate with two rows

```python
df_final=df3[df3.duplicated(subset=['Order ID','Products_mix'])]
df_final
```

```python
df_final=df3.groupby('Products_mix').count()
df_final.sort_values(by='Order ID',ascending=False)
```
